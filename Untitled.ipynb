### 1.1 Removing Non-Predictive Features

In this task, the dataset must first be cleaned by removing non-predictive or irrelevant features before handling missing data and normalization. The assignment instructions require us to “remove the non-predictive features and the potential target features from the dataset.” To determine which features should be removed, I referred both to the initial dataset exploration and to the guidelines covered in the lecture slides on data cleaning and feature preparation (CCE3503_02 Dataset Exploration) .

According to the slides, features should be removed when they:

Are identifiers that do not carry any predictive information

Have high cardinality or contain arbitrary codes (e.g., Ticket number in the Titanic example)

Contain irrelevant personal attributes

Contain excessive missing values that would introduce noise if imputed

Do not logically relate to the target variable

After inspecting all columns in the dataset, the following features were identified as non-predictive:

id — A row identifier. It carries no information about employee performance and is equivalent to dropping PassengerId as shown in the lecture slides.

zodiac_sign — Contains ~42% missing values and is not logically related to performance. This mirrors the case of Cabin in the slides, which was dropped for both irrelevance and high missingness.

favourite_colour_encoded — Encoded personal preference with no causal link to performance.

laptop_serial_number_last_digit — Arbitrary hardware metadata with no behavioural meaning.

parking_bay_number — Randomly assigned parking slot numbers; high-cardinality noise.

desk_location_zone_encoded — Arbitrary location code; similar to high-cardinality categorical features recommended for removal in the slides.

company_merch_size_encoded — Employees’ merchandise size (e.g., T-shirt size), which has no relevance to performance.

```python
cols_to_drop = [
    'id',
    'zodiac_sign',
    'favourite_colour_encoded',
    'laptop_serial_number_last_digit',
    'parking_bay_number',
    'desk_location_zone_encoded',
    'company_merch_size_encoded'
]

df = df.drop(columns=cols_to_drop)
df.info()
```

```python
# Identify numeric and categorical features
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Impute numeric features using median
from sklearn.impute import SimpleImputer

median_imputer = SimpleImputer(strategy='median')
df[numeric_cols] = median_imputer.fit_transform(df[numeric_cols])

# Impute categorical features (if any) using mode
mode_imputer = SimpleImputer(strategy='most_frequent')
df[categorical_cols] = mode_imputer.fit_transform(df[categorical_cols])

# Confirm no missing values remain
df.isna().sum().sum()
```

### 1.3 Normalizing the Data

The features in the dataset vary widely in their scales and units. Some variables are percentages (0–100), others measure counts, durations, or ratings on small integer scales. According to the lecture slides on feature scaling (CCE3503 — Dataset Exploration and Analysis, pages 33–37) , such differences in magnitude can cause features with larger numeric ranges to dominate those with smaller ranges during model training. Normalization is therefore required to ensure that all features contribute proportionally.

The slides present several scaling approaches, including Min–Max normalization and Standardization (zero mean and unit variance). For this assignment, Standardization using StandardScaler was chosen because:

It is the recommended scaling method for neural networks such as the MLPRegressor used in later tasks.

Standardization makes gradient-based optimisation more stable by centering each feature and scaling its variance to 1.

It is less sensitive to outliers than Min–Max scaling, which compresses all values into the 
[
0
,
1
]
[0,1] interval.

It directly follows the methodology demonstrated in the lecture slides using the StandardScaler class.

Only numerical features were standardized; categorical variables (sex and department) were left unchanged. After applying the scaler, all numerical features have zero mean and unit variance, making the dataset ready for feature selection and model training.

```python
from sklearn.preprocessing import StandardScaler

# Identify columns
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Standardize only numeric columns
scaler = StandardScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

df.head()
```

## 2.3 Train/Test Split

To evaluate the neural network model fairly, the dataset must be divided into a training set and a testing set. According to the Evaluation Metrics slides (CCE3503_05_evaluation.pdf, page 4) , a commonly recommended approach is to assign 80% of the data to training and 20% to testing. This ensures that the model has enough samples to learn general patterns, while still retaining a sufficiently large, unseen test set for evaluating generalisation performance.

Following this guideline, the dataset of 5000 samples was randomly split into an 80/20 ratio using scikit-learn’s train_test_split function. A fixed random seed (random_state=42) was applied to ensure reproducibility.

```python
from sklearn.model_selection import train_test_split

# Define feature matrix X and target vector y
X = df[selected_features]
y = df['performance_score']

# Perform the 80/20 train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,        # 20% test set as recommended in slides
    random_state=42,       # ensures reproducibility
    shuffle=True
)

# Show the resulting shapes
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

### 2.4 Neural Network Training and RMSE Evaluation

After selecting the features using the correlation-based filter method, a neural network model was trained to predict the employee performance score. According to the Machine Learning Models slides (CCE3503_07_models.pdf, pages 10–13) , a Multi-Layer Perceptron (MLP) can be used for both classification and regression tasks. It consists of interconnected neurons organised in layers, trained using backpropagation to learn complex, non-linear relationships.
For this reason, the MLPRegressor from scikit-learn was chosen as the prediction model.

Neural networks are sensitive to the scale of input features, and the slides highlight the importance of feature scaling prior to training. This was ensured earlier using standardisation, which places all numeric input variables on comparable scales and improves convergence stability for gradient-based optimisation.

The model was trained using 80% of the dataset and evaluated on the remaining 20%, following the recommendation in the Evaluation Metrics slides (CCE3503_05_evaluation.pdf, page 4) , which state that an 80/20 train/test split is commonly used to ensure sufficient training data while keeping a meaningful portion of unseen data for evaluation.

Model performance was measured using the Root Mean Square Error (RMSE), as presented in the regression metrics section of the Evaluation Metrics slides. RMSE is appropriate because it penalises larger prediction errors more strongly and is expressed in the same units as the target variable, making it interpretable. After training the MLPRegressor, the RMSE on the test set was computed from the model’s predictions.

```python
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error

# 1. Create the neural network model
mlp = MLPRegressor(
    hidden_layer_sizes=(100,),   # one hidden layer with 100 neurons (default)
    activation='relu',           # standard activation for MLPs
    solver='adam',               # recommended optimizer
    max_iter=1000,               # ensure convergence
    random_state=42              # reproducibility
)

# 2. Train the model
mlp.fit(X_train, y_train)

# 3. Predict on the test set
y_pred = mlp.predict(X_test)

# 4. Compute RMSE manually (because older sklearn lacks squared=False)
mse = mean_squared_error(y_test, y_pred)
rmse = mse ** 0.5
rmse
```

## Task 3 - Wrapper Methods

In this section we will use wrapper methods to evaluate subsets of features through training a model and measuring its performance. The wrapper methods used are; 

    - Sequential Forward Selection - builds up a subset one feature at a time
    - Sequential Backward Selection - starts with all features and removes them one at a time




### 3.1 Data Preparation for Wrapper Methods

Wrapper Methods require the full set of numeric features, as in task 2 the same 80/20 train test split will be used but excluding the target variable performance_score

```python
import mlxtend
from sklearn.model_selection import train_test_split

# 1. Use ALL numeric features (after cleaning & scaling)
wrapper_features = df.select_dtypes(include=['float64', 'int64']).columns.tolist()

# Remove the target from the feature list
wrapper_features.remove('performance_score')

# Feature matrix and target vector for wrapper methods
X_all = df[wrapper_features]
y_all = df['performance_score']

# 2. Train-test split (same settings as Task 2: 80/20, random_state=42)
X_train_wrap, X_test_wrap, y_train_wrap, y_test_wrap = train_test_split(
    X_all, y_all,
    test_size=0.20,
    random_state=42,
    shuffle=True
)

X_train_wrap.shape, X_test_wrap.shape, y_train_wrap.shape, y_test_wrap.shape
```

### 3.3 SFS 

SFS begins with empty feature set and then iteratively adds the feature that produces the largest improvement in performance. then it continues until it reaches the predetermined number of features which in this case is 40

```python
from sklearn.neural_network import MLPRegressor
from mlxtend.feature_selection import SequentialFeatureSelector as SFS

# Lightweight MLP for SFS
mlp_sfs = MLPRegressor(
    hidden_layer_sizes=(50,),
    activation='relu',
    solver='adam',
    max_iter=300,
    random_state=42
)

# SFS
sfs = SFS(
    estimator=mlp_sfs,
    k_features=40,
    forward=True,
    floating=False,
    scoring='neg_mean_squared_error',
    cv=3,
    n_jobs=-1
)

sfs = sfs.fit(X_train_wrap, y_train_wrap)
sfs_features = list(sfs.k_feature_names_)
sfs_features
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error

# Extract only SFS-selected features for train/test
X_train_sfs = X_train_wrap[sfs_features]
X_test_sfs = X_test_wrap[sfs_features]

# Full neural network (same as Task 2)
mlp_full = MLPRegressor(
    hidden_layer_sizes=(100,),   # same architecture as Task 2
    activation='relu',
    solver='adam',
    max_iter=1000,               # full convergence
    random_state=42
)

# Train the model
mlp_full.fit(X_train_sfs, y_train_wrap)

# Predict on test set
y_pred_sfs = mlp_full.predict(X_test_sfs)

# Compute RMSE (manual because of your sklearn version)
mse_sfs = mean_squared_error(y_test_wrap, y_pred_sfs)
rmse_sfs = mse_sfs ** 0.5

rmse_sfs
```

### 3.4 SBS

SBS begins with all features and them removes them one at a time, removing rhe feature which when removed will reduce least model performance. 


```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS

# SBS uses the same lightweight model as SFS
mlp_sbs = MLPRegressor(
    hidden_layer_sizes=(50,),
    activation='relu',
    solver='adam',
    max_iter=300,
    random_state=42
)

# Sequential Backward Selection
sbs = SFS(
    estimator=mlp_sbs,
    k_features=40,         # reduce to 40 features
    forward=False,         # THIS makes it SBS
    floating=False,
    scoring='neg_mean_squared_error',
    cv=3,
    n_jobs=-1
)

# Fit the SBS process
sbs = sbs.fit(X_train_wrap, y_train_wrap)

# Extract the selected features
sbs_features = list(sbs.k_feature_names_)
sbs_features
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error

# Extract SBS-selected features for train/test
X_train_sbs = X_train_wrap[sbs_features]
X_test_sbs = X_test_wrap[sbs_features]

# Full neural network (same architecture as Task 2)
mlp_full_sbs = MLPRegressor(
    hidden_layer_sizes=(100,),
    activation='relu',
    solver='adam',
    max_iter=1000,
    random_state=42
)

# Train the model
mlp_full_sbs.fit(X_train_sbs, y_train_wrap)

# Predict on test set
y_pred_sbs = mlp_full_sbs.predict(X_test_sbs)

# Compute RMSE
mse_sbs = mean_squared_error(y_test_wrap, y_pred_sbs)
rmse_sbs = mse_sbs ** 0.5

rmse_sbs
```

### 3.5 Strengths and limitations of SFS,SBS

Strengths of SFS 
    - good when irrelevant features exist, this is because it only adds features that improve  performance thus avoiding adding weak predictors
    - smaller search space as it starts with 0 features and thus is less computationally expensive than SBS 

Limitations of SFS
    - Once you add a feature in SFS, you cannot remove it and thus could be problematic later on 
    - slower than filter methods since it requires model training at every iteration 
    - risking local optima since early incorrect decisions could negatively impact final feature set 

Strengths of SBS 
    - the SBS method starts from the strongest full model which reduces risk of accidentally excluding important features 
    - removes features that contribute least to performance, hence making it robust 
    - because it reduces down from the full set it will preserve important features hence making it produce higher quality subsets 

Limitations of SBS
    - very expensive to compute since it evaluates models for every possible feature removal at every step 
    - similar to SFS, once a feature is removed it cannot be readded as it is irreversible
    - slower than SFS especially with larger feature spaces like in this context, took longer to run 


Both the SBS and SFS selected strong different subsets but SBS took longer to compute which is expected. Both have a lower rmse than the filter method highlighting their advantage in capturing

### Task 4 Feature Projection 

```python
from sklearn.model_selection import train_test_split

# Use all numeric features (after cleaning and scaling), exclude the target
pca_features = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
pca_features.remove('performance_score')

X_all = df[pca_features]
y_all = df['performance_score']

# Same 80/20 split as Task 2 (and Task 3)
X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(
    X_all, y_all,
    test_size=0.20,
    random_state=42,
    shuffle=True
)

X_train_pca.shape, X_test_pca.shape
```

```python
from sklearn.decomposition import PCA
import numpy as np

# Fit PCA on the training data (keep all components for now)
pca = PCA()
pca.fit(X_train_pca)

explained_var = pca.explained_variance_ratio_
explained_var[:10], explained_var.sum()
```

```python
n_features_original = X_train_pca.shape[1]
n_components_pca = pca.components_.shape[0]

n_features_original, n_components_pca, np.isclose(explained_var.sum(), 1.0)
```

```python
import matplotlib.pyplot as plt
import numpy as np

explained_var = pca.explained_variance_ratio_
cumulative_var = np.cumsum(explained_var)
components = np.arange(1, len(explained_var) + 1)

plt.figure(figsize=(12, 6))

# Scree plot (individual variance)
plt.bar(components, explained_var, alpha=0.7, label='Individual explained variance')

# Cumulative variance
plt.step(components, cumulative_var, where='mid', color='red', label='Cumulative explained variance')

plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('PCA – Explained Variance and Cumulative Variance')
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
pca_10 = PCA(n_components=10)
X_train_pca10 = pca_10.fit_transform(X_train_pca)
X_test_pca10 = pca_10.transform(X_test_pca)

X_train_pca10.shape, X_test_pca10.shape
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error
import numpy as np

# Define the neural network
mlp_pca = MLPRegressor(
    hidden_layer_sizes=(100,),   # same size as Task 2 & 3
    activation='relu',
    solver='adam',
    max_iter=1000,
    random_state=42
)

# Train on PCA-transformed features
mlp_pca.fit(X_train_pca10, y_train_pca)

# Predict on test PCA features
y_pred_pca = mlp_pca.predict(X_test_pca10)

# Compute RMSE
rmse_pca = np.sqrt(mean_squared_error(y_test_pca, y_pred_pca))
rmse_pca
```
