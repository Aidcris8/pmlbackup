### 1.1 Removing Non-Predictive Features

In this task, the dataset must first be cleaned by removing non-predictive or irrelevant features before handling missing data and normalization. The assignment instructions require us to “remove the non-predictive features and the potential target features from the dataset.” To determine which features should be removed, I referred both to the initial dataset exploration and to the guidelines covered in the lecture slides on data cleaning and feature preparation (CCE3503_02 Dataset Exploration) .

According to the slides, features should be removed when they:

Are identifiers that do not carry any predictive information

Have high cardinality or contain arbitrary codes (e.g., Ticket number in the Titanic example)

Contain irrelevant personal attributes

Contain excessive missing values that would introduce noise if imputed

Do not logically relate to the target variable

After inspecting all columns in the dataset, the following features were identified as non-predictive:

id — A row identifier. It carries no information about employee performance and is equivalent to dropping PassengerId as shown in the lecture slides.

zodiac_sign — Contains ~42% missing values and is not logically related to performance. This mirrors the case of Cabin in the slides, which was dropped for both irrelevance and high missingness.

favourite_colour_encoded — Encoded personal preference with no causal link to performance.

laptop_serial_number_last_digit — Arbitrary hardware metadata with no behavioural meaning.

parking_bay_number — Randomly assigned parking slot numbers; high-cardinality noise.

desk_location_zone_encoded — Arbitrary location code; similar to high-cardinality categorical features recommended for removal in the slides.

company_merch_size_encoded — Employees’ merchandise size (e.g., T-shirt size), which has no relevance to performance.

```python
cols_to_drop = [
    'id',                              # unique identifier
    'zodiac_sign',                     # >42% missing + superstition
    'favourite_colour_encoded',        # meaningless
    'average_coffees_per_day',         # obvious joke column
    'cafeteria_spend_avg_monthly',     # personal spending – unrelated
    'laptop_serial_number_last_digit', # random
    'parking_bay_number',              # random office allocation
    'desk_location_zone_encoded',      # random office layout
    'company_merch_size_encoded'       # irrelevant preference
]

df_clean = df.drop(columns=cols_to_drop).copy()
print(f"After removing {len(cols_to_drop)} irrelevant columns → {df_clean.shape}")
```

```python
# Verify only meaningful categorical columns remain
print("Remaining categorical columns:", df_clean.select_dtypes('object').columns.tolist())
```

```python
# Show missing values before imputation (for the report)
print("Missing values before imputation:")
print(df_clean.isnull().sum()[df_clean.isnull().sum() > 0].sort_values(ascending=False))
```

```python
# Median imputation for all numeric columns (robust + simple)
numeric_cols = df_clean.select_dtypes(include=['float64', 'int64']).columns
for col in numeric_cols:
    if df_clean[col].isnull().any():
        df_clean[col].fillna(df_clean[col].median(), inplace=True)

# Mode imputation for sex
df_clean['sex'].fillna(df_clean['sex'].mode()[0], inplace=True)

print(f"Missing values after imputation: {df_clean.isnull().sum().sum()}")  # Must be 0
```

```python
# One-hot encode sex and department (low cardinality → safe)
df_clean = pd.get_dummies(df_clean, columns=['sex', 'department'], drop_first=True)

print(f"After one-hot encoding → {df_clean.shape}")
# Expected: (5000, 88) → 85 + 1 (sex_M) + 4 (departments) - 2 dropped = +3 net
```

```python
# Separate features and target
y = df_clean['performance_score'].copy()
X = df_clean.drop('performance_score', axis=1)

print(f"Final X shape: {X.shape}")
print(f"Target y shape: {y.shape}")
```

```python
# Standardization – required for neural networks
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = pd.DataFrame(
    scaler.fit_transform(X),
    columns=X.columns,
    index=X.index
)

print("Standardization complete – all features now have mean ≈ 0 and std ≈ 1")
X_scaled.describe().loc[['mean', 'std']].round(6)
```

```python
# FINAL TASK 1 VALIDATION CHECK – Run this once at the end of Task 1
print("TASK 1 VALIDATION SUMMARY")
print("="*50)
print(f"1. Shape of final feature matrix X_scaled : {X_scaled.shape}")          # Should be (5000, 87)
print(f"2. Shape of target y                    : {y.shape}")                    # (5000,)
print(f"3. Any missing values in X_scaled?      : {X_scaled.isna().sum().sum()}")  # MUST be 0
print(f"4. Any missing values in y?             : {y.isna().sum()}")               # MUST be 0
print(f"5. Mean of all features (should ≈0)    : {X_scaled.mean().mean():.6f}")
print(f"6. Std of all features (should ≈1)     : {X_scaled.std().mean():.6f}")
print(f"7. Number of categorical columns left  : {len(['sex', 'department'])} → encoded into {X_scaled.select_dtypes('uint8').shape[1]} columns")
print(f"8. Junk columns still present?          : {'average_coffees_per_day' in X_scaled.columns or 'zodiac_sign' in X_scaled.columns or 'id' in X_scaled.columns}")

# Final verdict
if (X_scaled.shape == (5000, 87) and 
    X_scaled.isna().sum().sum() == 0 and 
    abs(X_scaled.mean().mean()) < 1e-10 and 
    abs(X_scaled.std().mean() - 1) < 0.01 and
    'average_coffees_per_day' not in X_scaled.columns):
    print("\nTASK 1 IS 100% CORRECT AND COMPLETE – READY FOR TASK 2!")
else:
    print("\nSomething is still wrong – check the numbers above.")
```

### 1.3 Normalizing the Data

The features in the dataset vary widely in their scales and units. Some variables are percentages (0–100), others measure counts, durations, or ratings on small integer scales. According to the lecture slides on feature scaling (CCE3503 — Dataset Exploration and Analysis, pages 33–37) , such differences in magnitude can cause features with larger numeric ranges to dominate those with smaller ranges during model training. Normalization is therefore required to ensure that all features contribute proportionally.

The slides present several scaling approaches, including Min–Max normalization and Standardization (zero mean and unit variance). For this assignment, Standardization using StandardScaler was chosen because:

It is the recommended scaling method for neural networks such as the MLPRegressor used in later tasks.

Standardization makes gradient-based optimisation more stable by centering each feature and scaling its variance to 1.

It is less sensitive to outliers than Min–Max scaling, which compresses all values into the 
[
0
,
1
]
[0,1] interval.

It directly follows the methodology demonstrated in the lecture slides using the StandardScaler class.

Only numerical features were standardized; categorical variables (sex and department) were left unchanged. After applying the scaler, all numerical features have zero mean and unit variance, making the dataset ready for feature selection and model training.

## Task 3 - Wrapper Methods

In this section we will use wrapper methods to evaluate subsets of features through training a model and measuring its performance. The wrapper methods used are; 

    - Sequential Forward Selection - builds up a subset one feature at a time
    - Sequential Backward Selection - starts with all features and removes them one at a time




### 3.1 Data Preparation for Wrapper Methods

Wrapper Methods require the full set of numeric features, as in task 2 the same 80/20 train test split will be used but excluding the target variable performance_score

```python
import numpy as np
from sklearn.model_selection import train_test_split

# Task 3 – Wrapper Method Setup

# 1. Use ALL preprocessed features (cleaned + imputed + encoded + scaled)
X_all = X_scaled.copy()   # shape should be (5000, 87)

# 2. Use the same target as in Task 2
y_all = y.copy()

# 3. Reproduce the exact same split as Task 2 USING X_all
# IMPORTANT: We must split based on X_all, not the 28-feature subset
X_train_wrap, X_test_wrap, y_train_wrap, y_test_wrap = train_test_split(
    X_all, y_all, test_size=0.2, random_state=42
)

# 4. Confirm shapes
X_train_wrap.shape, X_test_wrap.shape, y_train_wrap.shape, y_test_wrap.shape


```

### 3.3 SFS 

SFS begins with empty feature set and then iteratively adds the feature that produces the largest improvement in performance. then it continues until it reaches the predetermined number of features which in this case is 40

```python
from sklearn.linear_model import LinearRegression
from mlxtend.feature_selection import SequentialFeatureSelector as SFS

# 1. Use a fast estimator for wrapper selection (MLP is too slow for SFS)
lr = LinearRegression()

# 2. Configure SFS to select up to 40 features
sfs = SFS(
    estimator=lr,
    k_features=40,                  # Assignment requirement
    forward=True,                  # SFS (forward selection)
    floating=False,                # No floating steps
    scoring='neg_mean_squared_error',
    cv=3,                          # 3-fold cross-validation
    n_jobs=-1                      # Use all CPU cores
)

# 3. Fit SFS on the wrapper training set (87 standardized features)
sfs.fit(X_train_wrap, y_train_wrap)

# 4. Extract selected feature names
sfs_features = list(sfs.k_feature_names_)

# 5. Display list of selected SFS features
sfs_features
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error

# 1. Extract only SFS-selected features
X_train_sfs = X_train_wrap[sfs_features]
X_test_sfs = X_test_wrap[sfs_features]

# 2. Scale the selected features
scaler = StandardScaler()
X_train_sfs_scaled = scaler.fit_transform(X_train_sfs)
X_test_sfs_scaled = scaler.transform(X_test_sfs)

# 3. Train the SAME MLP architecture as Task 2 for fair comparison
mlp_sfs = MLPRegressor(
    hidden_layer_sizes=(100, 50),
    activation='relu',
    solver='adam',
    max_iter=1000,
    early_stopping=True,
    validation_fraction=0.1,
    n_iter_no_change=15,
    random_state=42
)

mlp_sfs.fit(X_train_sfs_scaled, y_train_wrap)

# 4. Predict on the test set
y_pred_sfs = mlp_sfs.predict(X_test_sfs_scaled)

# 5. Compute RMSE
rmse_sfs = (mean_squared_error(y_test_wrap, y_pred_sfs)) ** 0.5
rmse_sfs
```

### 3.4 SBS

SBS begins with all features and them removes them one at a time, removing rhe feature which when removed will reduce least model performance. 


```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS 
from sklearn.linear_model import LinearRegression 
# Same fast estimator for SBS
lr = LinearRegression()

sbs = SFS(
    estimator=lr,
    k_features=40,
    forward=False, # backward selection 
    floating=False,
    scoring='neg_mean_squared_error',
    cv=3,
    n_jobs=-1 
) 
# Fit SBS
sbs = sbs.fit(X_train_wrap, y_train_wrap)
# Extract selected feature names 
sbs_features = list(sbs.k_feature_names_) 
sbs_features
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error

# 1. Extract SBS-selected features for train/test
X_train_sbs = X_train_wrap[sbs_features]
X_test_sbs = X_test_wrap[sbs_features]

# 2. Scale the selected features (after feature selection)
scaler = StandardScaler()
X_train_sbs_scaled = scaler.fit_transform(X_train_sbs)
X_test_sbs_scaled = scaler.transform(X_test_sbs)

# 3. Train the SAME neural network architecture as Task 2 and SFS
mlp_sbs = MLPRegressor(
    hidden_layer_sizes=(100, 50),
    activation='relu',
    solver='adam',
    max_iter=1000,
    early_stopping=True,
    validation_fraction=0.1,
    n_iter_no_change=15,
    random_state=42
)

mlp_sbs.fit(X_train_sbs_scaled, y_train_wrap)

# 4. Predict on the test set
y_pred_sbs = mlp_sbs.predict(X_test_sbs_scaled)

# 5. Compute RMSE
mse_sbs = mean_squared_error(y_test_wrap, y_pred_sbs)
rmse_sbs = mse_sbs ** 0.5
rmse_sbs
```

### 3.5 Strengths and limitations of SFS,SBS

Strengths of SFS 
    - good when irrelevant features exist, this is because it only adds features that improve  performance thus avoiding adding weak predictors
    - smaller search space as it starts with 0 features and thus is less computationally expensive than SBS 

Limitations of SFS
    - Once you add a feature in SFS, you cannot remove it and thus could be problematic later on 
    - slower than filter methods since it requires model training at every iteration 
    - risking local optima since early incorrect decisions could negatively impact final feature set 

Strengths of SBS 
    - the SBS method starts from the strongest full model which reduces risk of accidentally excluding important features 
    - removes features that contribute least to performance, hence making it robust 
    - because it reduces down from the full set it will preserve important features hence making it produce higher quality subsets 

Limitations of SBS
    - very expensive to compute since it evaluates models for every possible feature removal at every step 
    - similar to SFS, once a feature is removed it cannot be readded as it is irreversible
    - slower than SFS especially with larger feature spaces like in this context, took longer to run 


Both the SBS and SFS selected strong different subsets but SBS took longer to compute which is expected. Both have a lower rmse than the filter method highlighting their advantage in capturing

```python
from sklearn.decomposition import PCA
import numpy as np

# 1. Fit PCA on the training data only (no leakage)
pca = PCA(random_state=42)   # reproducible PCA
pca.fit(X_train_pca)

# 2. Extract variance ratios
explained_var = pca.explained_variance_ratio_

# 3. Display the first few values and confirm total variance = 1.0
print("First 10 explained variance ratios:", explained_var[:10])
print("Total explained variance:", explained_var.sum())
```

```python
# PCA Diagnostic Information

n_features_original = X_train_pca.shape[1]
n_components_pca = pca.components_.shape[0]
variance_sum_correct = np.isclose(explained_var.sum(), 1.0)

print("Number of original features:", n_features_original)
print("Number of PCA components:", n_components_pca)
print("Does explained variance sum to 1.0?", variance_sum_correct)
```

```python
import matplotlib.pyplot as plt
import numpy as np

explained_var = pca.explained_variance_ratio_
cumulative_var = np.cumsum(explained_var)
components = np.arange(1, len(explained_var) + 1)

plt.figure(figsize=(12, 6))

# Scree plot (individual variance)
plt.bar(components, explained_var, alpha=0.7, label='Individual explained variance')

# Cumulative variance
plt.step(components, cumulative_var, where='mid', color='red', label='Cumulative explained variance')

plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('PCA – Explained Variance and Cumulative Variance')
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
cutoff = 0.95
cumulative_var = np.cumsum(explained_var)
n_components_cutoff = np.argmax(cumulative_var >= cutoff) + 1
n_components_cutoff
```

```python
pca_final = PCA(n_components=n_components_cutoff, random_state=42)

# Fit on training data only
X_train_pca_final = pca_final.fit_transform(X_train_pca)

# Transform test data
X_test_pca_final = pca_final.transform(X_test_pca)

X_train_pca_final.shape, X_test_pca_final.shape
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error
import numpy as np

# Neural network (same architecture as Task 2, SFS, and SBS)
mlp_pca = MLPRegressor(
    hidden_layer_sizes=(100, 50),
    activation='relu',
    solver='adam',
    max_iter=1000,
    early_stopping=True,
    validation_fraction=0.1,
    n_iter_no_change=15,
    random_state=42
)

# Train on PCA-transformed features
mlp_pca.fit(X_train_pca_final, y_train_pca)

# Predict on test PCA features
y_pred_pca = mlp_pca.predict(X_test_pca_final)

# Compute RMSE
rmse_pca = np.sqrt(mean_squared_error(y_test_pca, y_pred_pca))
rmse_pca
```
