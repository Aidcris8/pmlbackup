### Task 1 Problem Transformation



#### Task 1.1 Binary Relevance vs Classifier Chains

According to the multi-label classification techniques discussed in the lectures, problem transformation methods convert a multi-label problem into one or more single-label classification problems.

Binary Relevance (BR) is the simplest problem transformation approach. It decomposes a multi-label classification problem into multiple independent binary classification problems, one for each label. Each classifier is trained using only the original feature set, and label dependencies are ignored. While this approach is simple and computationally efficient, it assumes that all labels are independent, which may not be realistic in domains where labels are correlated.

Classifier Chains (CC) extend the binary relevance approach by explicitly modelling label dependencies. In classifier chains, binary classifiers are arranged in a sequence, where each classifier receives the original feature set together with the predictions of the preceding labels in the chain. This allows information about earlier predicted labels to influence later predictions, enabling the method to capture correlations between labels. However, this approach is sensitive to the order of labels and may suffer from error propagation if early predictions are incorrect.

#### 2.1 Hyperparameter Selection for the Neural Network

When selecting a suitable neural network for a multi-label classification task, several hyperparameters have a significant impact on model performance and generalisation. Based on the concepts discussed in the Models and Evaluation lectures, the following hyperparameters are selected for optimisation.

The number of hidden neurons (`hidden_layer_sizes`) controls the capacity of the neural network. As discussed in the Models lectures, increasing the number of neurons allows the model to learn more complex patterns, but also increases the risk of overfitting. Given the relatively high dimensionality of the dataset (103 features), optimising this parameter is important to balance underfitting and overfitting.

The regularisation parameter (`alpha`) determines the strength of L2 regularisation applied to the network weights. Regularisation is used to penalise overly complex models and improve generalisation, as covered in the lectures on model complexity and bias–variance trade-off. Optimising this parameter helps control overfitting, particularly in multi-label settings where multiple outputs are learned simultaneously.

The initial learning rate (`learning_rate_init`) influences how quickly and stably the neural network converges during training. As discussed in the Models lectures, an inappropriate learning rate can lead to unstable training or slow convergence. Optimising this parameter ensures efficient training and can improve final predictive performance.

These hyperparameters are optimised in later stages using cross-validation in order to select a neural network configuration that generalises well to unseen data.


#### 2.2 Hyperparameter Optimisation Technique

To optimise the selected neural network hyperparameters, the `RandomizedSearchCV` technique from the scikit-learn toolkit is used. Unlike exhaustive grid search, randomised search samples a fixed number of hyperparameter combinations from predefined distributions.

As discussed in the Evaluation lectures, randomised search is often more efficient than grid search when tuning multiple hyperparameters, particularly for models such as neural networks where training can be computationally expensive. It allows a broader exploration of the hyperparameter space while maintaining a manageable computational cost.

Given the number of hyperparameters being optimised and the size of the dataset, `RandomizedSearchCV` provides a good balance between search efficiency and performance, making it a suitable choice for this task.

## Task 3 — Performance Evaluation

### Metric Selection

To evaluate multi-label classification performance, metrics must account for the fact that each sample can have multiple labels simultaneously.

**Exact Match Ratio (Subset Accuracy)** is included to measure the proportion of samples for which the complete set of predicted labels matches the true set exactly. This metric is very strict and can be low in multi-label settings with many labels.

**Hamming Loss** measures the fraction of individual label predictions that are incorrect. It evaluates label-wise errors and is therefore informative when labels are sparse or when predicting the full label set exactly is difficult.

**Micro-averaged F1 score (F1-micro)** aggregates true positives, false positives, and false negatives across all labels before computing F1. This metric provides a balanced summary of precision and recall and is particularly appropriate when label frequencies are imbalanced.
